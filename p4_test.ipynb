```python
bits = ['32','48','48','16','8','32','32','128','128','16','16','16','16']
column_names = ['IN_PORT','EST_DST','EST_SRC','ETH_TYPE','IP_PROTO','IPv4_SRC','IPv4_DST','IPv6_SRC','IPv6_DST','TCP_SRC','TCP_DST','UDP_SRC','UDP_DST']
open_field_bits =  pd.DataFrame(columns = column_names)
open_field_bits.loc[0] = bits

open_table = pd.DataFrame(np.random.randint(low=0, high=2, size=(30, 13)),columns = ['IN_PORT','EST_DST','EST_SRC','ETH_TYPE','IP_PROTO','IPv4_SRC','IPv4_DST','IPv6_SRC','IPv6_DST','TCP_SRC','TCP_DST','UDP_SRC','UDP_DST'])


# print(open_field_bits)
# print(open_table)
```

```python
def find_unused(entry_row):
    unused_set = []
    for column in column_names:
        value = entry_row[column]
        if (value == 0):
            unused_set.append(column)
    return unused_set
```

```python
dfs = {}
# f_entry = open_table.loc[0]
# unused_field_key = find_unused(f_entry)

for idx, row in open_table.iterrows():
    unused_fields = find_unused(row)
    unused_fields_key = ','.join(unused_fields)
    
    if unused_fields_key in dfs:
        print(type(row))
        add_entry = row[row==1] 
        dfs[unused_fields_key].append(add_entry)
        print(add_entry)
    else:
        new_df = pd.DataFrame([row[row==1]],columns = unused_fields)
        dfs[unused_fields_key] = new_df
        print(add_entry)
# print (unused_fields)
print (dfs)
```
